```python
def twin(round):
    twinmatch = {"A" :"", "H" :""}
    twinround=[]
    for x in round:
        twinmatch["A"]=x['H']
        twinmatch["H"]=x['A']
        twinround.append(twinmatch.copy())        
    return twinround
```
